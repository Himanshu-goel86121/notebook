```python
import json
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd
from gensim.models import KeyedVectors
from gensim.scripts.glove2word2vec import glove2word2vec
from sklearn.metrics import pairwise_distances
import os
import keras
```

```python
from nltk.cluster.kmeans import KMeansClusterer
from scipy.cluster import  hierarchy
```

```python
file_location = "E:\Electronics_5_sample.csv"
w2v_location = "E:\embeddings\glove.6B.200d.txt"
temp_location = "E:\embeddings\glove.6B.200d_temp"
column = "reviewText"
```

```python
data = pd.read_csv(file_location).sample(frac=1.0).reset_index(drop=True).iloc[:20000]
```

```python
glove2word2vec(w2v_location, temp_location)
model = KeyedVectors.load_word2vec_format(temp_location)
os.remove(temp_location)
```

```python
data = data.dropna().reset_index(drop=True)
print(data.iloc[19835]["reviewText"])
print(data.iloc[19836]["reviewText"])
print(data["overall"].value_counts())
```

```python
sentences = data[column].values.tolist()
sentence_vectors = {}
for sentence in sentences:
    tokens = keras.preprocessing.text.text_to_word_sequence(sentence)
    sentence_vector = []
    if len(tokens) >= 2:
        for token in tokens:
            try:
                sentence_vector.append(model.wv[token])
            except:
                pass
        if len(sentence_vector) >=2:
            sentence_vectors[sentence] = np.mean(sentence_vector, axis=0).tolist()            
```

```python
print(len(sentence_vectors))
print(len(data))
```

```python
sentence_vectors_keys, sentence_vector_values = sentence_vectors.keys(), sentence_vectors.values()
clustering = KMeans(8).fit(np.array(list(sentence_vector_values)))
closest = pairwise_distances(clustering.cluster_centers_, np.array(list(sentence_vector_values))).argsort()
for i in closest:
    for j in i[:10]:
        print(data[j])
        print("*" * 100)
    print("#" * 100)
```

```python
sentence_vectors_keys, sentence_vector_values = sentence_vectors.keys(), sentence_vectors.values()
Z = hierarchy.linkage(np.array(list(sentence_vector_values)),"average", metric="cosine")
C = hierarchy.fcluster(Z, 0.010, criterion="distance")
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline


def tsne_plot_similar_words(title, labels, embedding_clusters, word_clusters, a, filename=None):
    plt.figure(figsize=(16, 9))
    colors = cm.rainbow(np.linspace(0, 1, len(labels)))
    for label, embeddings, words, color in zip(labels, embedding_clusters, word_clusters, colors):
        x = embeddings[:, 0]
        y = embeddings[:, 1]
        plt.scatter(x, y, c=color, alpha=a, label=label)
        for i, word in enumerate(words):
            plt.annotate(word, alpha=0.5, xy=(x[i], y[i]), xytext=(5, 2),
                         textcoords='offset points', ha='right', va='bottom', size  =8)
    plt.legend(loc=4)
    plt.title(title)
    plt.grid(True)
    if filename:
        plt.savefig(filename, format='png', dpi=150, bbox_inches='tight')
    plt.show()

tsne_plot_similar_words('Similar sentences', [str(i) for i in range(num_groups)], embeddings_en_2d, word_clusters, 0.7)
```
